```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Read the CSV file
data = pd.read_csv('/content/complete_dataset.csv')

# Filter the year from 1990 to 2020
filtered_years = data[(data['Year'] >= 1990) & (data['Year'] <= 2020)]['Year']
```

**East Asia and Pacific**

```python
East_Asia_Pacific_data = data[data['region'] == 'East Asia and Pacific']

EAP_columns_1 = ['Birth rate(pre 1000)', 'Child mortality rate %', 'Period life expectancy', 'Death rate']
East_Asia_Pacific_1 = East_Asia_Pacific_data[EAP_columns_1]

# Create a pairs plot
sns.pairplot(East_Asia_Pacific_1, height=2)
plt.show()
```

```python
East_Asia_Pacific_data = data[data['region'] == 'East Asia and Pacific']

EAP_columns_2 = ['Birth rate(pre 1000)', 'Consumer price index', 'female labor force participation rate', 'GDP', 'GDP per capita']
East_Asia_Pacific_2 = East_Asia_Pacific_data[EAP_columns_2]

# Create a pairs plot
sns.pairplot(East_Asia_Pacific_2, height=2)
plt.show()
```

```python
East_Asia_Pacific_data = data[data['region'] == 'East Asia and Pacific']

EAP_columns_3 = ['Birth rate(pre 1000)', 'Population', 'Rural population', 'Urban population']
East_Asia_Pacific_3 = East_Asia_Pacific_data[EAP_columns_3]

# Create a pairs plot
sns.pairplot(East_Asia_Pacific_3, height=2)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

columns_to_plot = ['Average years of education', 'Birth rate(pre 1000)']
East_Asia_Pacific_subset = East_Asia_Pacific_data[columns_to_plot]

# Create a scatter plot
sns.scatterplot(x='Average years of education', y='Birth rate(pre 1000)', data=East_Asia_Pacific_subset)
plt.xlabel('Average years of education')
plt.ylabel('Birth rate(pre 1000)')
plt.title('Scatter Plot: Education vs Birth Rate')
plt.show()
```
