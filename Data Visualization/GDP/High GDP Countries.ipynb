**20 counyries with highest GDP**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('/content/complete_dataset.csv')

# Filter the year from 1990 to 2020
filtered_data = data[(data['Year'] >= 1990) & (data['Year'] <= 2020)]

# Get the top 20 countries with the highest GDP
top_20_gdp = filtered_data.groupby('Entity')['GDP'].max().nlargest(20).index.tolist()

# Filter data for the top 20 countries
top_20_data = filtered_data[filtered_data['Entity'].isin(top_20_gdp)]

# Select the columns for comparison
columns_to_compare = ['Birth rate(pre 1000)', 'Child mortality rate %', 'Period life expectancy', 'Death rate']

# Create a pair plot
sns.pairplot(top_20_data, vars=columns_to_compare, hue='Entity')
plt.show()
```
