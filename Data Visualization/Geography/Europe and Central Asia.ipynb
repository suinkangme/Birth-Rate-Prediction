```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Read the CSV file
data = pd.read_csv('/content/complete_dataset.csv')

# Filter the year from 1990 to 2020
filtered_years = data[(data['Year'] >= 1990) & (data['Year'] <= 2020)]['Year']
```

**Europe and Central Asia**

```python
Europe_Central_Asia_data = data[data['region'] == 'Europe and Central Asia']

ECA_columns_1 = ['Birth rate(pre 1000)', 'Child mortality rate %', 'Period life expectancy', 'Death rate']
Europe_Central_Asia_1 = Europe_Central_Asia_data[ECA_columns_1]

# Create a pairs plot
sns.pairplot(Europe_Central_Asia_1, height=2)
plt.show()
```

```python
Europe_Central_Asia_data = data[data['region'] == 'Europe and Central Asia']

ECA_columns_2 = ['Birth rate(pre 1000)', 'Consumer price index', 'female labor force participation rate', 'GDP', 'GDP per capita']
Europe_Central_Asia_2 = Europe_Central_Asia_data[ECA_columns_2]

# Create a pairs plot
sns.pairplot(Europe_Central_Asia_2, height=2)
plt.show()
```

```python
Europe_Central_Asia_data = data[data['region'] == 'Europe and Central Asia']

ECA_columns_3 = ['Birth rate(pre 1000)', 'Population', 'Rural population', 'Urban population']
Europe_Central_Asia_3 = Europe_Central_Asia_data[ECA_columns_3]

# Create a pairs plot
sns.pairplot(Europe_Central_Asia_3, height=2)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

columns_to_plot = ['Average years of education', 'Birth rate(pre 1000)']
Europe_Central_Asia_subset = Europe_Central_Asia_data[columns_to_plot]

# Create a scatter plot
sns.scatterplot(x='Average years of education', y='Birth rate(pre 1000)', data=Europe_Central_Asia_subset)
plt.xlabel('Average years of education')
plt.ylabel('Birth rate(pre 1000)')
plt.title('Scatter Plot: Education vs Birth Rate')
plt.show()
```
